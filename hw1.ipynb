```python
#1.2 

# Define the system of differential equations
def system(state, alpha):
    x, v = state
    dxdt = v
    dvdt = -x - alpha * x**3
    return np.array([dxdt, dvdt])

# Define the range for x and v
x = np.linspace(-2, 2, 400)
v = np.linspace(-2, 2, 400)

# Create a meshgrid
X, V = np.meshgrid(x, v)

# Compute the derivatives
alpha_values = [1, -1]
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

for i, alpha in enumerate(alpha_values):
    DXDT = V
    DVDT = -X - alpha * X**3

    # Plot the phase portrait
    axes[i].streamplot(X, V, DXDT, DVDT, color='b')
    axes[i].set_title(f'Phase Portrait for $\\alpha = {alpha}$')
    axes[i].set_xlabel('$x$')
    axes[i].set_ylabel('$\dot{x}$')
    axes[i].axhline(0, color='black', linewidth=0.5)
    axes[i].axvline(0, color='black', linewidth=0.5)
    axes[i].grid(color='gray', linestyle='--', linewidth=0.5)

plt.tight_layout()
plt.show()
```

```python
#1.2


# Define the function f(x, alpha) = -x - alpha * x^3
def f(x, alpha):
    return -x - alpha * x**3

# Set up the plot range
x_vals = np.linspace(-2, 2, 400)
alpha_vals = np.linspace(-10, 10, 400)
X, A = np.meshgrid(x_vals, alpha_vals)

# Compute f(x, alpha)
# alpha < 0 set A = -A
f_vals = f(-A,X)

# Compute f(x, alpha) > 0 (regions where the function is positive)
positive_region = f_vals > 0

# Plotting the equilibrium curves (f(x, alpha) = 0)
plt.figure(figsize=(8, 6))
plt.contour(X, A, f_vals, levels=[0], colors='black', linewidths=2)

# Shade the regions where f(x, alpha) > 0
plt.contourf(X, A, positive_region, levels=[0.5, 1], colors=['lightblue'], alpha=0.5)

# Labels and titles
plt.xlabel(r'$\alpha$')
plt.ylabel(r'$x$')
plt.title(r'Equilibrium Curve and Stability Analysis for $-x - \alpha x^3 = 0$')

# Show the plot
plt.grid(True)
plt.show()
```

```python
#1.6

# Define the potential energy function
def V(x):
    if x < -1:
        return x + 1
    elif -1 <= x <= 1:
        return 0
    else:
        return -x + 1

# Define the range for x
x = np.linspace(-2, 2, 400)
V_values = np.array([V(xi) for xi in x])

# Plot the potential energy function
plt.figure(figsize=(10, 6))
plt.plot(x, V_values, label='$V(x)$')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.xlabel('$x$')
plt.ylabel('$V(x)$')
plt.title('Potential Energy Function $V(x)$')
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.legend()
plt.show()
```

```python
#1.6

# Define the potential energy function and its derivative
def V(x):
    if x < -1:
        return x + 1
    elif -1 <= x <= 1:
        return 0
    else:
        return -x + 1

def dV_dx(x):
    if x < -1:
        return 1
    elif -1 <= x <= 1:
        return 0
    else:
        return -1

# Define the range for x
x = np.linspace(-2, 2, 400)

# Define the range for y
y = np.linspace(-2, 2, 400)

# Create a grid for the streamplot
X, Y = np.meshgrid(x, y)

# Define the vector field
U = Y  # Velocity field: U = y
V = np.array([[dV_dx(xi) for xi in row] for row in X])  # Use the derivative of the potential

# Plot the phase paths
plt.figure(figsize=(10, 6))
plt.streamplot(X, Y, U, V, color='red', linewidth=1, density=1.5, arrowsize=1.5)

# Add labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Phase Paths for V(x)')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.show()
```

```python
#1.11

# Define parameters
omega = 0.5  #frequency 

# Define the range for x and y
x = np.linspace(-3*np.pi, 3*np.pi, 400)
y = np.linspace(-3, 3, 400)

# Create a grid for the streamplot
X, Y = np.meshgrid(x, y)

# Define the derivatives for the original pendulum equation
U1 = Y
V1 = -omega**2 * np.sin(X)

# Define the derivatives for the approximated equation
U2 = Y
V2 = -omega**2 * (X - (1/6) * X**3)

# Plot the phase paths for the original pendulum equation
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.streamplot(X, Y, U1, V1, color='red', linewidth=1, density=1.5, arrowsize=1.5)
plt.scatter([n*np.pi for n in range(-3, 4)], [0]*7, color='blue', zorder=5, label='Equilibrium points')
plt.xlabel('x')
plt.ylabel('y ')
plt.title('Phase Diagram for $\ddot{x} + \omega^2 \sin(x) = 0$')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.legend()

# Plot the phase paths for the approximated equation
plt.subplot(1, 2, 2)
plt.streamplot(X, Y, U2, V2, color='red', linewidth=1, density=1.5, arrowsize=1.5)
plt.scatter([0, np.sqrt(6), -np.sqrt(6)], [0, 0, 0], color='blue', zorder=5, label='Equilibrium points')
plt.xlabel('x')
plt.ylabel('y ')
plt.title('Phase Diagram for $\ddot{x} + \omega^2 ( x - \\frac{1}{6} x^3 ) = 0$')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.legend()

plt.tight_layout()
plt.show()
```

```python
#1.18


# Define the function f(x, lambda) = (x - lambda)(x^2 - lambda)
def f(x, lam):
    return (x - lam) * (x**2 - lam)

# Set up the plot range
x_vals = np.linspace(-2, 2, 400)
lambda_vals = np.linspace(-1, 2, 400)
X, L = np.meshgrid(x_vals, lambda_vals)

# Compute f(x, lambda)
f_vals = f(X, L)

# Compute f(x, lambda) > 0 (regions where the function is positive)
positive_region = f_vals > 0

# Plotting the equilibrium curves (f(x, lambda) = 0)
plt.figure(figsize=(8, 6))
plt.contour(L, X, f_vals, levels=[0], colors='black', linewidths=2)

# Shade the regions where f(x, lambda) > 0
plt.contourf(L, X, positive_region, levels=[0.5, 1], colors=['lightblue'], alpha=0.5)

# Labels and titles
plt.xlabel(r'$\lambda$')
plt.ylabel(r'$x$')
plt.title(r'Equilibrium Curve and Stability Analysis for $(x - \lambda)(x^2 - \lambda) = 0$')

# Show the plot
plt.grid(True)
plt.show()
```

```python
#1.21
# Define a function to plot phase diagrams in subplots
def plot_phase_diagram(ax, U_func, V_func, title):
    # Define the range for x and y
    x = np.linspace(-5, 5, 400)
    y = np.linspace(-5, 5, 400)

    # Create a grid for the streamplot
    X, Y = np.meshgrid(x, y)

    # Define the vector field
    U = U_func(X, Y)  # x-velocity field
    V = V_func(X, Y)  # y-velocity field

    # Plot the phase paths on the given axis
    ax.streamplot(X, Y, U, V, color='red', linewidth=1, density=1.5, arrowsize=1.5)

    # Add labels and title
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title(title)
    ax.axhline(0, color='black', linewidth=0.5)
    ax.axvline(0, color='black', linewidth=0.5)
    ax.grid(color='gray', linestyle='--', linewidth=0.5)

# Create subplots
fig, axes = plt.subplots(1, 3, figsize=(18, 6))  # 1 row, 3 columns

# (i) dx/dt = y, dy/dt = 0
U_func1 = lambda X, Y: Y
V_func1 = lambda X, Y: np.zeros_like(X)  # Ensure this is an array, not a scalar
plot_phase_diagram(axes[0], U_func1, V_func1, r'Phase Diagram: $\dot{x} = y, \dot{y} = 0$')

# (ii) dx/dt = y, dy/dt = 1
U_func2 = lambda X, Y: Y
V_func2 = lambda X, Y: np.ones_like(X)  # Ensure this is an array
plot_phase_diagram(axes[1], U_func2, V_func2, r'Phase Diagram: $\dot{x} = y, \dot{y} = 1$')

# (iii) dx/dt = y, dy/dt = y
U_func3 = lambda X, Y: Y
V_func3 = lambda X, Y: Y  # This already returns an array
plot_phase_diagram(axes[2], U_func3, V_func3, r'Phase Diagram: $\dot{x} = y, \dot{y} = y$')

# Adjust layout
plt.tight_layout()
plt.show()
```
